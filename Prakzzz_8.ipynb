```python
import string

# Caesar Cipher
def caesar_encrypt(teks, shift=3):
    hasil = ""
    for c in teks:
        if c.isalpha():
            start = ord('A') if c.isupper() else ord('a')
            hasil += chr((ord(c) - start + shift) % 26 + start)
        else:
            hasil += c
    return hasil

def caesar_decrypt(teks, shift=3):
    return caesar_encrypt(teks, -shift)


# Vigenere Cipher
def vigenere_encrypt(teks, key):
    hasil = ""
    key = key.lower()
    idx = 0

    for c in teks:
        if c.isalpha():
            shift = ord(key[idx % len(key)]) - ord('a')
            base = ord('A') if c.isupper() else ord('a')
            hasil += chr((ord(c) - base + shift) % 26 + base)
            idx += 1
        else:
            hasil += c
    return hasil


def vigenere_decrypt(teks, key):
    hasil = ""
    key = key.lower()
    idx = 0

    for c in teks:
        if c.isalpha():
            shift = ord(key[idx % len(key)]) - ord('a')
            base = ord('A') if c.isupper() else ord('a')
            hasil += chr((ord(c) - base - shift) % 26 + base)
            idx += 1
        else:
            hasil += c
    return hasil


def process(teks, mode):
    if mode == 1:
        shift = int(input("Shift berapa? "))
        return caesar_encrypt(teks, shift)

    elif mode == 2:
        shift = int(input("Shift berapa untuk decrypt? "))
        return caesar_decrypt(teks, shift)
    
    elif mode == 3:
        key = input("Masukkan key Vigenere: ")
        return vigenere_encrypt(teks, key)

    elif mode == 4:
        key = input("Masukkan key Vigenere: ")
        return vigenere_decrypt(teks, key)

print("Word Transformation - Encryption & Decryption")
print("1. Caesar Encrypt")
print("2. Caesar Decrypt")
print("3. Vigenere Encrypt")
print("4. Vigenere Decrypt")

teks = input("Masukkan teks: ")
mode = int(input("Pilih mode 1-4: "))

hasil = process(teks, mode)
print("\nHasil:", hasil)
```

```python
from collections import defaultdict
import string

def mapper(teks):
    """Memecah teks menjadi pasangan (huruf, 1)"""
    mapped = []
    for c in teks.lower():
        if c in string.ascii_lowercase:    # hanya huruf alfabet
            mapped.append((c, 1))
    return mapped

def shuffle(mapped_data):
    """Mengelompokkan key yang sama (mirip Hadoop shuffle phase)"""
    shuffled = defaultdict(list)
    for key, value in mapped_data:
        shuffled[key].append(value)
    return shuffled

def reducer(shuffled_data):
    """Menjumlahkan nilai dari setiap huruf"""
    reduced = {}
    for key, values in shuffled_data.items():
        reduced[key] = sum(values)
    return reduced

def hadoop_freq_count(teks):
    mapped = mapper(teks)
    shuffled = shuffle(mapped)
    reduced = reducer(shuffled)
    return dict(sorted(reduced.items()))
    
teks = input("Masukkan teks: ")

hasil = hadoop_freq_count(teks)

print("\nFrekuensi huruf:")
for huruf, jumlah in hasil.items():
    print(f"{huruf}: {jumlah}")
```

```python
import re

def bersihkan(teks):
    """Menghapus non-huruf dan lowercase"""
    return re.sub(r'[^A-Za-z]', '', teks).lower()


def palindrome_similarity(teks):
    clean = bersihkan(teks)
    n = len(clean)
    
    if n == 0:
        return 0
    
    cocok = 0
    total = n
    
    # Bandingkan tiap huruf dengan pasangannya dari belakang
    for i in range(n):
        if clean[i] == clean[n - 1 - i]:
            cocok += 1
    
    similarity = (cocok / total) * 100
    return round(similarity, 2), clean

teks = input("Masukkan teks: ")

similarity, cleaned = palindrome_similarity(teks)

print("\nTeks yang diproses:", cleaned)
print(f"Palindrome Similarity: {similarity}%")
```
